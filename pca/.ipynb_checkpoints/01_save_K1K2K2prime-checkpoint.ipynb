```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
```

```python
# data table
table=[]
T=0.5
for T in [0.1, 0.2, 0.5, 1.0]:
    for U in [0.5, 1.0, 2.0, 3.0]:
        for e in [-2.0, -1.5, -1.0, -0.5, 0.0, 0.5, 1.0, 1.5, 2.0]:
            table.append((U,T,e))

    remove_from_table=[(3.0, 0.2, 0.5), (3.0, 0.2, -0.5), (3.0, 0.1, 0.5), (3.0, 0.1, -0.5)]
    for to_remove in remove_from_table:
        if to_remove in table:
            table.remove(to_remove)

data_size=len(table)
print(data_size)
```

```python
#explore datasets
U = 2.0
T = 0.1
e = 0.0
filename = f"/mnt/home/jzang/ceph/freq_frg/SIAM.KF.U{U}.T{T}.e{e}.h5"
def explore_hdf5_group(group, prefix=""):
    for key in group.keys():
        item = group[key]
        item_path = f"{prefix}/{key}" if prefix else key
        if isinstance(item, h5py.Group):
            print(f"Group: {item_path} ({len(item)} members)")
            explore_hdf5_group(item, item_path)
        elif isinstance(item, h5py.Dataset):
            print(f"Dataset: {item_path}, Shape: {item.shape}, Dtype: {item.dtype}")
with h5py.File(filename, 'r') as file:
    explore_hdf5_group(file)
```
