```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
```

```python
# data table
table=[]
for T in [0.1, 0.2, 0.5, 1.0]:
    for U in [0.5, 1.0, 2.0, 3.0]:
        for e in [-2.0, -1.5, -1.0, -0.5, 0.0, 0.5, 1.0, 1.5, 2.0]:
            table.append((U,T,e))

    remove_from_table=[(3.0, 0.2, 0.5), (3.0, 0.2, -0.5), (3.0, 0.1, 0.5), (3.0, 0.1, -0.5)]
    for to_remove in remove_from_table:
        if to_remove in table:
            table.remove(to_remove)

data_size=len(table)
print(data_size)
```

```python
for i, (U,T,e) in enumerate(table):
    if U==2.0 and e==0.0:
        filename = f"/mnt/home/jzang/ceph/freq_frg/SIAM.KF.U{U}.T{T}.e{e}.h5"
        with h5py.File(filename, 'r') as file:
            mesh_bosonic=np.array(file['vertex_total']['mesh_bosonic'])
            vertex=np.array(file['vertex_total']['data_uddu'])

            plt.plot(mesh_bosonic,-vertex[:, 0, 0, 1, 1, 1, 0].imag/U**2,label=f"T={T}")
plt.xlim(-5,5)
plt.legend()
plt.xlabel(r"$\omega$")

plt.title(r"$\chi_{M}^{R}(\omega)=-\Gamma_{\uparrow \downarrow \downarrow \uparrow}^{1222}(-20,-20, \omega) / U^{2}$, U=2, e=0.0")
plt.show()
```

```python
#load and save total vertex with uddu component
vertex_uddu=np.zeros((data_size,201, 101, 101, 2, 2, 2, 2),dtype=np.complex128)

for i, (U,T,e) in enumerate(table):
    # print(U,T,e)
    filename = f"/mnt/home/jzang/ceph/freq_frg/SIAM.KF.U{U}.T{T}.e{e}.h5"
    with h5py.File(filename, 'r') as file:
        vertex_uddu[i]=np.array(file['vertex_total']['data_uddu'])
vertex_uddu_mean=np.mean(vertex_uddu,axis=0)
vertex_uddu=vertex_uddu-vertex_uddu_mean
filename = f"/mnt/home/jzang/ceph/freq_frg/store_data/vertex_total_uddu.h5"
with h5py.File(filename, 'a') as file:
    file['vertex_uddu']=vertex_uddu
    file['vertex_uddu_mean']=vertex_uddu_mean
    file['parameters']=table
    
```
